```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import requests
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
monthlyInflation=pd.read_csv('inflationCPI_OECD_monthly.csv')
monthlyInflation
usefulMonthlyInflation=monthlyInflation[['LOCATION','TIME','Value']]
```

```python
url ='https://raw.githubusercontent.com/tm19784/tm19784.github.io/main/jsonFolder/rawPriceBTC.json'
html = requests.get(url)
```

```python
Bitcoin_price_json = html.json()
BTC=pd.DataFrame(Bitcoin_price_json)
BTC_useful=BTC[['time_close','price_close']]
BTC_useful
```

```python
BTC_useful['time_close'] = pd.to_datetime(BTC_useful['time_close'])
BTC_useful
BTC_monthly = BTC_useful.set_index('time_close').resample('1M').mean().reset_index()
BTC_monthly
```

```python
usefulMonthlyInflation['TIME'] = pd.to_datetime(usefulMonthlyInflation['TIME'])
```

```python
Ger_usefulMonthlyInflation = usefulMonthlyInflation[usefulMonthlyInflation['LOCATION'].str.contains('DEU')]
Ger_usefulMonthlyInflation.reset_index()
```

```python
Ger_BTC = Ger_usefulMonthlyInflation[-78:].reset_index()[['LOCATION', 'TIME', 'Value']].join(BTC_monthly)
Ger_BTC
```

```python
Ger_BTC.rename(columns={'Value': 'Germany_inflation_rate'}, inplace=True)
Ger_BTC.rename(columns={'price_close': 'BTC_closing_price'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set()
```

```python
Ger_BTC_correlation = Ger_BTC[['Germany_inflation_rate','BTC_closing_price']]
Ger_BTC.to_csv('Ger_BTC.csv')
Germany_Bitcoin_correlation = pd.read_csv('Ger_BTC.csv')
```

```python
sns.regplot(x='BTC_closing_price',y='Germany_inflation_rate',data=Germany_Bitcoin_correlation).set(title='Regression of closing price of Ethereum on inflation rate of Japan')
```

```python
from scipy.stats import linregress
```

```python
slope, intercept, rvalue, pvalue, stderr = linregress(x=Germany_Bitcoin_correlation.dropna()['BTC_closing_price'], y=Germany_Bitcoin_correlation.dropna()['Germany_inflation_rate'])
```

```python
regression_equation = print('y = ',np.round(slope,2),' * x + ',np.round(intercept,2))
regression_R2 = print('RÂ² = ',np.round(rvalue**2,2))
```

```python
x = Germany_Bitcoin_correlation.dropna()['BTC_closing_price']
y = Germany_Bitcoin_correlation.dropna()['Germany_inflation_rate']

poly_model = make_pipeline(PolynomialFeatures(8),
                           LinearRegression())
poly_model.fit(x[:, np.newaxis], y)
xfit = np.linspace(min(x), max(x), 1000)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit,color='crimson');
plt.xlabel('Ethereum Closing Price')
plt.ylabel('Japanese Inflation Rate')
plt.title('Relationship between Ethereum price and Japanese inflation')
```

```python
df_fitz = pd.DataFrame(xfit,yfit)
df_fitz.rename({'index':'yfit'}, inplace=True)
df_fitz_2 = df_fitz.reset_index()
df_fitz_2.rename(columns={'index':'yfit'}, inplace=True)
df_fitz_2.rename(columns={0:'xfit'}, inplace=True)
df_fitz_2
```

```python
Regression_output_table = pd.concat([Germany_Bitcoin_correlation, df_fitz_2])
Regression_output_table.head(85)
```

```python
Regression_output_table.to_csv('inflation_Regression_output_Ger_BTC.csv')
```
