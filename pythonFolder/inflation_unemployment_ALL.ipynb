```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import requests
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
yearlyInflation=pd.read_csv('inflationWorldBank_yearly_uncleaned.csv')
yearlyInflation
usefulYearlyInflation=yearlyInflation[['Country Name','2015', '2016', '2017', '2018', '2019', '2020']]
# Here we upload yearly inflation data from the World Bank, and then only include columns in the data that countain the name of the country, as well as the years since the Bitcoin data begins.
```

```python
ALL_usefulYearlyInflation = usefulYearlyInflation[usefulYearlyInflation['Country Name'].str.contains('Switzerland|India|Nigeria|Peru|Philippines|Turkey|Vietnam|United States|China|Japan|Germany')]
bestALL_usefulYearlyInflation = ALL_usefulYearlyInflation.set_index('Country Name').T
bestALL_usefulYearlyInflation.index.name = 'Year'
bestALL_usefulYearlyInflation
# From the previous dataframe, we keep only observations that consist of the countries that are part of the list that have the highest percentage of their citizens owning cryptocurrencies.
# The '|' operater represents 'or' separating the possible strings, hence why this is used.
# Setting the country as the index and then tranposing puts each column as a separate country.
# We then label the index 'Year' as the index will need a name in order to manipulate in Vega.
```

```python
bestALL_usefulYearlyInflation.to_csv('inflation_ALLcountries.csv')
# This is then exported as a CSV file that I upload to my GitHub website.
# Below I do exactly the same but with yearly unemployment rates (from the same dataset).
```
