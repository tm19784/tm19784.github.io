```python
url ='https://raw.githubusercontent.com/tm19784/tm19784.github.io/main/yahooFinanceScraped.html'
html = requests.get(url)
```

```python
soup = BS(html.content, 'html.parser')
soup
# We now have the part of the HTML that we want to scrape.
```

```python
countries=soup.findAll('b')
countries
countries_list=[i.text.replace(':','') for i in countries]
countries_list
```

```python
percentages=soup.findAll('span')
percentages
percentages_list=[int(i.text.replace('%','')) for i in percentages]
percentages_list
```

```python
df_Yahoo=pd.DataFrame([countries_list,percentages_list]).T
df_Yahoo.rename(columns={0: 'Country or Area'}, inplace=True)
df_Yahoo.rename(columns={1: 'Citizens (%)'}, inplace=True)
df_Yahoo.sort_values('Country or Area', ascending=True, inplace=True)
df_Yahoo_better = df_Yahoo.reset_index()[['Country or Area', 'Citizens (%)']]
# Here I have isolated the part of the edited HTML that is useful to me, and then completed minor cleaning.
# This ensures that all observations are left as plain words and integers.
# I also renamed the columns so the code would be easier to follow later on.
# In addition, I also put the countries in alphabetical order as this is the format of the other dataframe.
```

```python
url_Continent = 'https://statisticstimes.com/geography/countries-by-continents.php'
# To display more information on my bar chart, I am also scraping a website that categorises countries by continent.
```

```python
html = requests.get(url_Continent)
Continent_dfs=pd.read_html(html.content)
Continent_dfs[2]
# The third table on the page is the one that I want, hence I need to select the 2nd dataframe.
```

```python
usefulContinent = Continent_dfs[2][['Country or Area','Continent', 'ISO-alpha3 Code']]
finishedContinent = usefulContinent[usefulContinent['ISO-alpha3 Code'].str.contains('VNM|USA|NGA|CHE|DEU|PHL|PER|CHN|IND|JPN|TUR', na=False)]
finishedContinent['Country or Area'] = finishedContinent['Country or Area'].str.replace('Viet Nam', 'Vietnam').str.replace('United States of America', 'U.S.')
cleanedContinent = finishedContinent[['Country or Area', 'Continent']].reset_index()
cleanedContinent = cleanedContinent[['Country or Area', 'Continent']]
cleanedContinent
# Here I keep rows from the original dataframe that have country codes matching the countries from the other dataframe.
# I also rename two country names so they match that of the other dataset.
```

```python
df_Yahoo_less = df_Yahoo_better[['Citizens (%)']]
df_Yahoo_continent = cleanedContinent.join(df_Yahoo_less)
df_Yahoo_continent
# This merges the dataset by the name of the country.
# I can now use the 'Continent' column to display extra information in Vega.
```

```python
df_Yahoo_continent.to_csv('top10_Crypto_Countries.csv')
# This exports our above dataframe as a CSV file.
```
