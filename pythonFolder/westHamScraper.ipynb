```python
import requests, pandas as pd, numpy as np
```

```python
url='https://www.whufc.com/fixtures/first-team/fixtures-and-results?field_competition=36&field_competition_season=6535&submit-button=Apply'
```

```python
html = requests.get(url)
whufc_dfs=pd.read_html(html.content)
whufc_dfs
# This creates a list of dataframes for all the covered variables; each dataframe represents a different month.
```

```python
whufc_df=pd.concat(whufc_dfs).replace('> Match Centre',np.nan).dropna(how='all',axis=0).dropna(how='all',axis=1)
whufc_df.columns=['date','stadium','home','home_goals','x','away_goals','away']
whufc_df=whufc_df.drop('x',axis=1)
whufc_df.reset_index()
# This creates a big dataframe by concatenating the list of dataframes, removing irrelevant information (such as '> Match Centre'), dropping missing values, and fitting columns for the data that we want.
```

```python
whufc_df_useful = whufc_df.reset_index()[0:17]
whufc_df_useful
```

```python
whufc_df_useful.to_csv('goalsWestHamStats.csv')
# We then drop the rows from the dataframe that involve matches that haven't happened yet, and then export as a CSV.
```
