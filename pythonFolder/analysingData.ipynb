```python
no1s = no1s.assign(danceable = no1s['danceability'] > 0.5)
# This assigns 'true' values in a new 'danceable' column if danceability is greater than 0.5 and zero otherwise...
```

```python
dance = no1s[no1s['danceable'] == True]
no_dance = no1s[no1s['danceable'] ==  False]
plt.plot(dance.index, dance['energy'], 'r.', label='Danceable')
plt.plot(no_dance.index, no_dance['energy'], 'bx', label='Not Danceable')
```
